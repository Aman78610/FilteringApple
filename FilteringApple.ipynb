```python
kernelx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
kernely = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
#kernelnew = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
img_prewittx = cv2.filter2D(grayImage, -1, kernelx)
img_prewitty = cv2.filter2D(grayImage, -1, kernely)
#img_prewittz = cv2.filter2D(grayImage, -1, kernelnew)

        #filter2D(src, dst, ddepth , kernel, anchor, delta, BORDER_DEFAULT );
        # ddepth = -1, means destination image has depth same as input image.

#Diagonal
kernelR = np.array([[0,1,1],[-1,0,1],[-1,-1,0]])
kernelL = np.array([[0,-1,-1],[1,0,-1],[1,1,0]])
img_prewittR = cv2.filter2D(grayImage, -1, kernelR)
img_prewittL = cv2.filter2D(grayImage, -1, kernelL)
```
